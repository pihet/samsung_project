```python
import urllib.request
import json

# 발급받으신 키를 그대로 사용합니다.
client_id = "72zv8pmji6"
client_secret = "70xiSyIythvjoa3JBpRmEcqssrFFD7cQziEZJ0mH"

# 1. API 호출 URL을 NAVER API HUB 주소로 설정
url = "https://naverapihub.apigw.ntruss.com/shopping/v1/categories"

# 2. 요청 Body 설정 (쇼핑인사이트는 POST 요청이므로 Body 값이 필요합니다)
body = {
    "startDate": "2026-07-01",
    "endDate": "2026-07-20",
    "timeUnit": "date",
    "category": [
        {"name": "패션의류", "param": ["50000000"]} # 카테고리 이름 및 코드
    ]
}
data = json.dumps(body).encode("utf-8")

request = urllib.request.Request(url, data=data)

# 헤더 이름을 모두 소문자로 변경
request.add_header("x-ncp-apigw-api-key-id", client_id)
request.add_header("x-ncp-apigw-api-key", client_secret)
request.add_header("content-type", "application/json")


try:
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if rescode == 200:
        response_body = response.read()
        print(response_body.decode('utf-8'))
    else:
        print("Error Code:" + str(rescode))
except urllib.error.HTTPError as e:
    print("HTTP Error:", e.code)
    print(e.read().decode('utf-8'))
```
